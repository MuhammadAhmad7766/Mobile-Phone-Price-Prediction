```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB, GaussianNB, MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')
```

```python

# 1. Read CSV
filename = 'mobile phone price prediction.csv'
data = pd.read_csv(filename)
```

```python
# ==========================================
# LAB 10: Data Pre-Processing
# ==========================================
print("\n--- LAB 10 TASKS ---")

# 7. Dropping unnecessary columns (Unnamed: 0 is an index column)
if 'Unnamed: 0' in data.columns:
    data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# SPECIFIC TASK: Type conversion
# The 'Price' column is an object with commas (e.g., "9,999"). 
# We clean it and convert to Int as per Lab 10 requirements.
if data['Price'].dtype == 'object':
    data['Price'] = data['Price'].str.replace(',', '')
    data['Price'] = data['Price'].astype(np.int64)
    print("\n'Price' column converted to Int64.")
```

```python
# 9. Splitting into X and Y
# Based on Lab 11 snippet "y = data.iloc[:,-1]", the target is the last column ('Processor_name').
X = data.iloc[:, 0:-1]
Y = data.iloc[:, -1]

print(f"\nX Shape: {X.shape}")
print(f"Y Shape: {Y.shape}")
```

```python
# 10. Converting Object columns into Int columns (Factorization)
cat_columns = X.select_dtypes(['object']).columns
X[cat_columns] = X[cat_columns].apply(lambda x: pd.factorize(x)[0])

print("Object columns in X have been factorized to integers.")

```
