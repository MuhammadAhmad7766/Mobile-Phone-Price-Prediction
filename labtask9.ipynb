```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB, GaussianNB, MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')
```

```python
print("--- LAB 9 TASKS ---")

# 1. Read CSV
filename = 'mobile phone price prediction.csv'
data = pd.read_csv(filename)
```

```python
# 2. Print Top 5 rows
print("\nTop 5 rows:")
print(data.head())
```

```python

# 3. Print Bottom 5 rows
print("\nBottom 5 rows:")
print(data.tail())
```

```python
# 4. Print Number of Rows and Columns
print(f"\nRows: {data.shape[0]}")
print(f"Columns: {data.shape[1]}")
```

```python

# 5. Determine null columns
print("\nNull Values per column:")
print(data.isnull().sum())
```

```python
# 6. Filling null values
# Using Mode for categorical/object columns and Mean for numerical columns
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = data[col].fillna(data[col].mode()[0])
    else:
        data[col] = data[col].fillna(data[col].mean())
```
